```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_suicide_rate'
```

```python
page1=requests.get(url)
page1
```

```python
soup=BeautifulSoup(page1.text)
soup
```

```python
table=soup.find('div',class_="mw-parser-output")
```

```python
table = table.find_all('table')
```

```python
table_data =[]
for i in table:
    table_data.append(i.find_all('td'))
```

```python
table_data = table_data[-1]
```

```python
all_data = []
for i in table_data:
    all_data.append(i.text.replace('\n',''))
all_data
```

```python
data = all_data
```

```python
rank = []
states = []
s_rate20 = []
s_rate15 = []

i = 0

while i < len(data):
    rank.append(data[i])
    i += 4
    
i = 1

while i < len(data):
    states.append(data[i])
    i += 4

i = 2

while i < len(data):
    s_rate20.append(data[i])
    i += 4
    
i = 3

while i < len(data):
    s_rate15.append(data[i])
    i += 4
```

```python
dictionary = {
    'Rank': rank,
    'States': states,
    'Suicide Rate (Per 1 Lakh) 2020': s_rate20,
    'Suicide Rate (Per 1 Lakh) 2015': s_rate15,
}
```

```python
dictionary
```

```python
suicide = pd.DataFrame(dictionary)
```

```python
suicide.to_csv('Suicidal Rates.csv',index=False)
```
